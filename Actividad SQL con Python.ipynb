# SQL con Python

```python
import numpy as np
import pandas as pd
import sqlite3 # <-- Interactuar con BD SQLite
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = sqlite3.connect("hr.sqlite"); conn
```

## Parte 1

- Mostrar los nombres (first_name, last_name) utilizando el alias “first name”, “last name” de la tabla de empleados


```python
sql = """
SELECT first_name AS 'first name', last_name AS 'last name'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el “department ID” único de la tabla de empleados

```python
df1 = pd.read_sql_query("SELECT DISTINCT department_id FROM employees", conn); 
df1
```

- Obtener todos los detalles de los empleados de la tabla de empleados ordenados por nombre, de manera descendente


```python
sql = """
SELECT *
FROM employees 
ORDER BY first_name ASC
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido, salario y la remuneración de todos los empleados (la remuneración se calcula como el 12% del salario).


```python
sql = """
SELECT first_name AS 'nombre', last_name AS 'apellido', salary AS 'sueldo', salary*0.12 AS 'remuneración'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el  “employee ID”, nombre, apellido y salario ordenados de manera ascendente por el salario.


```python
df1 = pd.read_sql_query("SELECT employee_id, first_name, last_name, salary FROM employees ORDER BY salary ASC", conn); 
df1
```

- Obtener el total del salario pagado a los empleados.


```python
sql = """
SELECT round(sum(salary), 2) AS 'Salario total'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el salario máximo y mínimo de la tabla de empleados.


```python
sql = """
SELECT min(salary) AS 'salario mínimo', max(salary) AS 'salario máximo'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el salario medio y el número de empleados de la tabla de empleados.


```python
df1 = pd.read_sql_query("SELECT avg(salary) AS 'salario medio', count(employee_id) AS 'número de empleados' FROM employees", conn); 
df1
```

- Obtener el número de puestos de trabajo disponibles en la tabla de empleados.

```python
sql = """
SELECT COUNT (DISTINCT job_id) AS 'número de puestos de trabajo disponibles'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

## Parte 2

- Obtener todos los nombres de la tabla de empleados en mayúsculas.


```python
df1 = pd.read_sql_query("SELECT UPPER(first_name) AS 'nombre del empleado' FROM employees", conn); 
df1
```

- Obtener los 3 primeros caracteres del nombre de cada empleado.


```python
df1 = pd.read_sql_query("SELECT SUBSTRING(first_name, 1, 3) AS 'primeras tres letras del nombre' FROM employees", conn); 
df1
```

- Obtener el nombre completo de todos los empleados de la tabla de empleados.

```python
sql = """
SELECT (first_name || ' ' || last_name) AS 'nombre completo'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el salario mensual (a 2 decimales) de todos los empleados.


```python
df1 = pd.read_sql_query("SELECT round(salary, 2) AS 'salario mensual' FROM employees", conn); 
df1
```

- Obtener nombre, apellido y salario de todos los empleados cuyo salario no esté en el rango de 10,000 a 15,000.


```python
sql = """
SELECT first_name, last_name, salary 
FROM employees 
WHERE salary > 15000
OR salary < 10000
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido y departamento de todos los empleados de los departamentos 30 o 100 en orden alfabético por “department ID”.


```python
sql = """
SELECT em.first_name, em.last_name, em.department_id, dep.depart_name
FROM employees em
INNER JOIN departments dep ON dep.department_id = em.department_id
WHERE em.department_id = 30
OR em.department_id = 100
ORDER BY dep.depart_name ASC
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido y salario de todos los empleados cuyo salario no esté en el rango de 10,000 a 15,000 y estén en el departamento 30 o 100.


```python
sql = """
SELECT first_name, last_name, department_id, salary
FROM employees 
WHERE (department_id = 30 OR department_id = 100)
AND (salary > 15000 OR salary < 10000)
"""
pd.read_sql_query(sql, conn)
```
