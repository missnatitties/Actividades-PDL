# SQL con Python

```python
import numpy as np
import pandas as pd
import sqlite3 # <-- Interactuar con BD SQLite
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = sqlite3.connect("hr.sqlite"); conn
```

## Parte 1

- Mostrar los nombres (first_name, last_name) utilizando el alias “first name”, “last name” de la tabla de empleados


```python
sql = """
SELECT first_name AS 'first name', last_name AS 'last name'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el “department ID” único de la tabla de empleados

```python
df1 = pd.read_sql_query("SELECT DISTINCT department_id FROM employees", conn); 
df1
```

- Obtener todos los detalles de los empleados de la tabla de empleados ordenados por nombre, de manera descendente


```python
sql = """
SELECT *
FROM employees 
ORDER BY first_name ASC
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido, salario y la remuneración de todos los empleados (la remuneración se calcula como el 12% del salario).


```python
sql = """
SELECT first_name AS 'nombre', last_name AS 'apellido', salary AS 'sueldo', salary*0.12 AS 'remuneración'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el  “employee ID”, nombre, apellido y salario ordenados de manera ascendente por el salario.


```python
df1 = pd.read_sql_query("SELECT employee_id, first_name, last_name, salary FROM employees ORDER BY salary ASC", conn); 
df1
```

- Obtener el total del salario pagado a los empleados.


```python
sql = """
SELECT round(sum(salary), 2) AS 'Salario total'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el salario máximo y mínimo de la tabla de empleados.


```python
sql = """
SELECT min(salary) AS 'salario mínimo', max(salary) AS 'salario máximo'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el salario medio y el número de empleados de la tabla de empleados.


```python
df1 = pd.read_sql_query("SELECT avg(salary) AS 'salario medio', count(employee_id) AS 'número de empleados' FROM employees", conn); 
df1
```

- Obtener el número de puestos de trabajo disponibles en la tabla de empleados.

```python
sql = """
SELECT COUNT (DISTINCT job_id) AS 'número de puestos de trabajo disponibles'
FROM employees 
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido y salario de todos los empleados cuyo salario no esté en el rango de 10,000 a 15,000.


```python
sql = """
SELECT first_name, last_name, salary 
FROM employees 
WHERE salary > 15000
OR salary < 10000
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido y departamento de todos los empleados de los departamentos 30 o 100 en orden alfabético por “department ID”.


```python
sql = """
SELECT em.first_name, em.last_name, em.department_id, dep.depart_name
FROM employees em
INNER JOIN departments dep ON dep.department_id = em.department_id
WHERE em.department_id = 30
OR em.department_id = 100
ORDER BY dep.depart_name ASC
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido y salario de todos los empleados cuyo salario no esté en el rango de 10,000 a 15,000 y estén en el departamento 30 o 100.


```python
sql = """
SELECT first_name, last_name, department_id, salary
FROM employees 
WHERE (department_id = 30 OR department_id = 100)
AND (salary > 15000 OR salary < 10000)
"""
pd.read_sql_query(sql, conn)
```

## Parte 3

- Obtener el apellido, el cargo y el salario de todos los empleados cuyo trabajo es el de “Programmer” o “Shipping Clerk”, y cuyo salario no sea igual a 4,500, 10,000 o 15,000 dólares.


```python
sql = """
SELECT em.last_name, job.job_title, em.salary
FROM employees em 
INNER JOIN jobs job ON job.job_id = em.job_id
WHERE (job_title = 'Programmer' OR job_title = 'Shipping Clerk')
AND (salary != 4500 OR salary != 10000 OR salary != 15000)
"""
pd.read_sql_query(sql, conn)
```

- Obtener los apellidos de los empleados cuyos nombres tienen exactamente 6 caracteres.


```python
df1 = pd.read_sql_query("SELECT last_name FROM employees WHERE LENGTH(first_name) = 6", conn); 
df1
```

- Obtener los apellidos de los empleados que tienen "e" como tercer carácter.


```python
df1 = pd.read_sql_query("SELECT last_name FROM employees WHERE substring(last_name, 3, 1) = 'e'", conn); 
df1
```

- Obtener el salario máximo de un empleado que trabaja como “Programmer”.

```python
sql = """
SELECT max(em.salary) AS 'salario máximo', job.job_title AS 'puesto'
FROM employees em 
INNER JOIN jobs job ON job.job_id = em.job_id
WHERE job_title = 'Programmer' 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el salario medio y el número de empleados que trabajan en el departamento 90.


```python
sql = """
SELECT avg(salary) AS 'salario medio', count(employee_id) AS 'número de empleados', department_id AS 'departamento'
FROM employees 
WHERE department_id = 90 
"""
pd.read_sql_query(sql, conn)
```

- Obtener el salario más alto, el más bajo, la suma y la media de todos los salarios.


```python
df1 = pd.read_sql_query("SELECT max(salary), min(salary), sum(salary), round(avg(salary), 2) FROM employees", conn); 
df1
```

- Obtener el número de empleados con el mismo puesto.


```python
sql = """
SELECT count(em.employee_id) AS 'número de empleados', j.job_title AS 'puesto'
FROM employees em
INNER JOIN jobs j ON j.job_id = em.job_id
GROUP BY j.job_title
ORDER BY count(em.employee_id) DESC
"""
pd.read_sql_query(sql, conn)
```

- Obtener la diferencia entre el salario más alto y el más bajo.


```python
df1 = pd.read_sql_query("SELECT max(salary)-min(salary) AS 'diferencia entre salario máximo y mínimo' FROM employees", conn); 
df1
```

- Obtener el “manager ID” y el salario del empleado con menor sueldo reportando a ese “manager''.

## Parte 4

- Obtener el “department ID” y el total de salarios a pagar en cada departamento.



```python
df1 = pd.read_sql_query("SELECT department_id, sum(salary) FROM employees GROUP BY department_id", conn); 
df1
```

- Obtener el salario medio de cada ID de puesto, excluyendo “Programmer”.


```python
sql = """
SELECT avg(em.salary) AS 'salario medio', j.job_title AS 'puesto'
FROM employees em
INNER JOIN jobs j ON j.job_id = em.job_id
WHERE j.job_title != 'Programmer'
GROUP BY j.job_title
"""
pd.read_sql_query(sql, conn)
```

- Obtener el job ID y el salario máximo de los empleados que su salario máximo es mayor o igual a $4,000.


```python
df1 = pd.read_sql_query("SELECT job_id, max(salary) FROM employees WHERE salary >= 4000 GROUP BY job_id", conn); 
df1
```

- Obtener el salario medio de todos los departamentos que emplean a más de 10 trabajadores.


```python
sql = """
SELECT department_id, avg(salary), COUNT(employee_id) AS 'conteo'
FROM employees 
GROUP BY department_id
HAVING COUNT(employee_id) > 10
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido y  salario de los empleados que tienen un salario más alto que el empleado cuyo apellido es “Bull”.

- Obtener nombre y apellido de los empleados que tienen un “manager” que trabaja en un departamento con sede en EUA.

## Parte 5

- Obtener nombre, apellido y salario de los empleados cuyo salario es igual al salario mínimo para su rol.


```python
sql = """
SELECT em.first_name, em.last_name, em.salary, j.job_title, j.min_salary 
FROM employees em 
INNER JOIN jobs j ON j.job_id = em.job_id
WHERE em.salary = j.min_salary
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido y salario de los empleados que ganan más que el salario medio y que trabajan en cualquier área de IT.



```python
sql = """
SELECT em.first_name, em.last_name, em.salary, dep.depart_name
FROM employees em 
INNER JOIN departments dep ON dep.department_id = em.department_id
WHERE dep.depart_name = 'IT'
AND em.salary > (SELECT avg(salary) FROM employees)
"""
pd.read_sql_query(sql, conn)
```

- Obtener nombre, apellido y salario de los empleados que ganan más que el Sr. Bell.


```python
df1 = pd.read_sql_query("SELECT first_name, last_name, salary FROM employees WHERE salary > (SELECT salary FROM employees WHERE last_name = 'Bell')", conn); 
df1
```

- Obtener nombre y apellido de los empleados que no son “supervisors”.


```python
sql = """
SELECT em.first_name, em.last_name
FROM employees em 
INNER JOIN jobs j ON j.job_id = em.job_id
WHERE j.job_id != 'Supervisor'
"""
pd.read_sql_query(sql, conn)
```

- Obtener el employee ID, nombre, apellido y salario de todos los empleados cuyo salario es superior a la media de sus departamentos.


```python
sql = """
SELECT em.employee_id, em.first_name, em.last_name, em.salary
FROM employees em 
LEFT JOIN (SELECT avg(salary) AS 'salario_ref', department_id FROM employees ref GROUP BY department_id) AS ref
ON ref.department_id = em.department_id
WHERE em.salary > salario_ref
"""
pd.read_sql_query(sql, conn)

```

- Obtener el 5º salario máximo de la tabla de empleados.


```python
sql = """
SELECT salary 
FROM employees  
ORDER BY salary DESC
LIMIT 1 OFFSET 4;
"""
pd.read_sql_query(sql, conn)
```

- Obtener las direcciones (location_id, street_address, city, state_province, country_name) de todos los departamentos.


```python
sql = """
SELECT d.depart_name, l.location_id, l.street_address, l.city, l.state_province, c.country_name 
FROM departments d  
INNER JOIN locations l ON l.location_id = d.location_id
INNER JOIN countries c ON c.country_id = l.country_id
"""
pd.read_sql_query(sql, conn)
```

- Obtener el “employee ID”, nombre y apellido de cada empleado junto con el nombre, apellido y “manager ID” de su “manager”.


```python
sql = """
SELECT em.employee_id AS 'ID del empleado', em.first_name AS 'nombre del empleado', em.last_name AS 'apellido del empleado', ref.first_name AS 'nombre del gerente', ref.last_name AS 'apellido del gerente', em.manager_id as 'ID del gerente'
FROM employees em 
LEFT JOIN (SELECT first_name, last_name, employee_id FROM employees) AS ref
ON ref.employee_id = em.manager_id
"""
pd.read_sql_query(sql, conn)
```

- Obtener el job_title y el salario medio de los empleados.

```python
sql = """
SELECT j.job_title AS 'puesto', avg(em.salary) AS 'salario medio'
FROM employees em
INNER JOIN jobs j ON j.job_id = em.job_id
GROUP BY j.job_id
"""
pd.read_sql_query(sql, conn)
```
