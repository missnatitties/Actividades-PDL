# Parte 3

Mostrar las primeras 21 imágenes en una malla de 3 filas x 7 columnas

No mostrar los ejes

Mostrar el título de cada video (10 caracteres)

Mostrar las imágenes en blanco y negro

En caso de que la foto no exista, mostrar el siguiente “Warning” (advertencia) a color: http://www.dagdrivarn.se/BILDER/stambilder/MISSING.jpg


```python
df_pictures = df[0:21]
df_pictures.head()
```

```python
links = list(df_pictures['thumbnail_link']) #Listado de los links
fig, axs = plt.subplots(3, 7, figsize = (12,4)) #Cuántos renglones y columnas + de qué tamaño serán tus imágenes
titles = list(df_pictures['title']) #Listado de los títulos del video

for ax, url, title in zip(axs.flatten(), links, titles):
    try:
        ax.imshow(io.imread(url)[:,:,0], cmap = 'gray') #Con el [:,:,0]: los ':' permiten mantener el tamaño que ya habías preestablecido y el 0 permite cambiar a escala de grises vía cmap
        ax.set_title(title[0:10]) #Número de caracteres mostrados en el título
        ax.axis('off') #Quitas los ejes
    except:
        ax.imshow(io.imread('http://www.dagdrivarn.se/BILDER/stambilder/MISSING.jpg')) #Muestras la imagen de 'Missing' cuando no haya imagen disponible para el video
        ax.set_title(title[0:10])
        ax.axis('off')
```
