# Parte 4

Mostrar la columna “tags” (etiquetas)

Los tags están separados por una barra vertical

Crear una nueva columna con el nombre “tag_conteo” que enumere los tags (etiquetas) que contiene cada video

“sports|mexico|news" tendría 3 tags

```python
df['tags']
```

```python
df['tag_conteo'] = df['tags'].apply(lambda x:x.count('|') + 1 if x!='[none]' else 0)

df.head()
```

# Parte 5

Crear una nueva columna con nombre “Likes”

Likes contendrá los booleanos “True” y “False”

Será “True” si el número de Likes supera al número de “Dislikes”

En caso contrario será “False”

```python
df['preference'] = df.apply(lambda x:True if x.likes > x.dislikes else False, axis = 1) #El 1 es para realizar la operación por columna
df.head()
```

# Parte 6

Convertir el archivo .json en un dataframe

Revisar el “antes" y “después" en las imágenes


```python
import json 
with open('./Files actividades/US_category_id.json', 'r') as f:data = json.loads(f.read())
data
```

```python
dataframe = pd.json_normalize(data = data, record_path = ['items'])
dataframe.head()
```

```python
dataframe.rename(columns={'id': 'category_id', 'snippet.title': 'category_title'}, inplace = True)
dataframe[['category_id', 'category_title']].head()
```

# Parte 7

Combinar el dataframe original (.csv) con el nuevo dataframe (.json)

```python
df.head(1)
```

```python
dataframe.head(1)
```

```python
df['category_id'] = df['category_id'].astype('int')
dataframe['category_id'] = dataframe['category_id'].astype('int')
```

```python
merged_df = pd.merge(df, dataframe, on = 'category_id')
merged_df = merged_df[['title', 'channel_title', 'category_title', 'views', 'likes', 'dislikes', 'comment_count', 'description', 'tag_conteo', 'preference']]
merged_df.head()
```

# Parte 12

Crea una malla de 1 fila y 2 columnas con dos gráficos de puntos en escala logarítmica
- “Likes” vs. “Dislikes” en color rosa, semi transparente con marcadores tipo “+” con título y etiquetas en ambos ejes
- “Views” vs. “Comments” en color azul, semi transparente, con marcadores tipo “*” con título y etiquetas en ambos ejes


```python
fig, axs = plt.subplots(1,2, figsize = (20,6))
axs[0].scatter(df['likes'], df['dislikes'], color = 'magenta', marker = '+', alpha = 0.2)
axs[0].set_title('Likes vs. dislikes')
axs[0].set_xlabel('Likes')
axs[0].set_ylabel('Dislikes')
axs[0].set_xscale('log')
axs[0].set_yscale('log')

axs[1].scatter(df['views'], df['comment_count'], color = 'blue', marker = '*', alpha = 0.2)
axs[1].set_title('Views vs. comments')
axs[1].set_xlabel('Views')
axs[1].set_ylabel('Comments')
axs[1].set_xscale('log')
axs[1].set_yscale('log')
```
