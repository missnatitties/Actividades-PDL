```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
X = pd.DataFrame(data.data, columns=data.feature_names); X.head()
```

```python
y = data.target; y[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
parameters = {'n_estimators': range(100, 200, 25),
              'criterion': ['gini', 'entropy'],
              'max_depth': list(range(1, 5)) + [None],
              'max_features': ['auto', 'sqrt', 'log2']
             }
model = RandomForestClassifier()
gs = GridSearchCV(model, parameters, cv=5, n_jobs=-1, verbose=1)
gs.fit(X_train, y_train)
gs.best_score_, gs.best_params_
```

```python
final = RandomForestClassifier(**gs.best_params_).fit(X_train, y_train)
```

```python
final.score(X_test, y_test)
```
