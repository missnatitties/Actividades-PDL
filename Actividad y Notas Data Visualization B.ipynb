# Stacked graphs 

```python
fig = plt.figure(figsize=(12,4))

t = np.arange(0.0,1.0,0.1)
s = np.sin(2*np.pi*t)

ax1 = fig.add_subplot(3,1,1) #índice 3,1,1
ax1.set_ylabel('Voltaje')
ax1.set_title('Curva de PE')
line1 = ax1.plot(t,s + 5.0, color='red', lw=5)
line2 = ax1.plot(t,s + 3.0, color='blue', lw=3)
line3 = ax1.plot(t,s, color='pink', lw=3)

ax2 = fig.add_subplot(3,1,2) #índice 3,1,2
ax2.set_ylabel('Voltaje')
ax2.set_title('Curva de PE')
line1 = ax2.plot(t,s + 5.0, color='red', lw=5, linestyle='dashed')
line2 = ax2.plot(t,s + 3.0, color='blue', lw=3, linestyle='-.')
line3 = ax2.plot(t,s, color='brown', lw=3, linestyle=':')

ax3 = fig.add_subplot(3,1,3) #índice 3,1,3
ax3.set_ylabel('Voltaje')
ax3.set_title('Curva de PE')
line1 = ax3.plot(t,s + 5.0, color='red', lw=5, marker='+')
line2 = ax3.plot(t,s + 3.0, color='green', lw=3, marker='o')
line3 = ax3.plot(t,s, color='orange', lw=1, marker='^')

plt.subplots_adjust(wspace=0.5, hspace=0.3)
```

# Utilizando estilos predefinidos

```python
fig = plt.figure(figsize=(6,6))
with plt.style.context('dark_background'):
    N = 4
    x = np.linspace(0,N)
    ncolors = len(plt.rcParams['axes.prop_cycle'])
    shift = np.linspace(0, N, ncolors, endpoint = False)
    for l in shift:
        plt.plot(x,np.sin(x+l), '--')
```

```python
fig = plt.figure(figsize=(6,6))
with plt.style.context('bmh'):
    N = 4
    x = np.linspace(0,N)
    ncolors = len(plt.rcParams['axes.prop_cycle'])
    shift = np.linspace(0, N, ncolors, endpoint = False)
    for l in shift:
        plt.plot(x,np.sin(x+l), '--')
```

```python
import pandas as pd
import seaborn as sns
x = np.arange(0,4,0.1)
y = np.sin(x)

df = pd.DataFrame({'x':x,'sinx':y,'cosx':np.cos(x)})
df = df.set_index('x')

grid = sns.lmplot(x='x', y='sinx', data=df.reset_index())
grid.axes[0,0].hlines(0,-1,5) #Estableces una línea hotizontal en 0 que va de -1 a 5
ax = grid.axes
```

```python
sns.set(style='dark')

rs = np.random.RandomState(4)
pos = rs.randint(-1,2,(20,5)).cumsum(axis=1)
step = np.tile(range(5),20)
walk = np.repeat(range(20),5)
df = pd.DataFrame(np.c_[pos.flat, step, walk], columns=['pos','step','walk'])

grid = sns.FacetGrid(df, col='walk', hue='walk', col_wrap=5, height=3)

grid.map(plt.axhline, y=0, ls=':') #Línea en y = 0

grid.map(plt.plot, 'step', 'pos', marker='o', ms=4) #Línea en círculo
```

# Paso 4

```python
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
import plotly.express as px

import pandas as pd

df = pd.read_csv('C:/Users/ncordova/Downloads/auto-mpg.csv')
df['hp'] = df['horsepower']
df['cyl'] = df['cylinders']
df = df[df.hp != "?"]
df["hp"] = df["hp"].astype(int)

app = dash.Dash(__name__)

app.layout = html.Div(children=[
    html.H1(children='Histograma', style={'color':'black'}),
    dcc.Graph(id="Grafico_horsepower"),
    dcc.Dropdown(
        id='cylinder-dropdown',
        options=[{'label':i, 'value':i} for i in df['cyl'].unique()],
        value= '4'
        ),
])

@app.callback(
    Output('Grafico_horsepower', 'figure'),
    Input('cylinder-dropdown', 'value')
)
def update_figure(selected_cylinder):
    filtered_df = df[df.cyl == selected_cylinder]

    fig= px.histogram(filtered_df, x='hp')

    fig.update_layout(transition_duration=400)
    
    return fig

if __name__ == '__main__':
    app.run_server(debug=True,use_reloader=False)
```

# Paso 5

```python
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
import plotly.express as px

import pandas as pd

df = pd.read_csv('C:/Users/ncordova/Downloads/auto-mpg.csv')

df["mpg"] = df["mpg"].astype(int)
a=df.iloc[:,8]
df['name'] = a.str.split(' ').str[0]

app = dash.Dash(__name__)

app.layout = html.Div(children=[
    html.H1(children='Violin_graph', style={'color':'black'}),
    dcc.Graph(id="graph_mpg"),
    dcc.Dropdown(
        id='car_name_dropdown',
        options=[
            {'label': 'ford', 'value': 'ford'},
            {'label': 'chevrolet', 'value': 'chevrolet'},
            {'label': 'toyota', 'value': 'toyota'}
        ],
        value= 'ford'
        ),
])

@app.callback(
    Output('graph_mpg', 'figure'),
    Input('car_name_dropdown', 'value')
)
def update_figure(selected_car):
    filtered_df = df[df.name == selected_car]
    
    fig= px.violin(filtered_df, y="mpg", x="name", color="name", box=True, points="all", hover_data=df.columns)
    
    fig.update_layout(transition_duration=500)
    
    return fig

if __name__ == '__main__':
   app.run_server(debug=True, use_reloader=False)
```
