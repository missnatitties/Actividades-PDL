```python
data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
# hay muchos hiperparámetros para probar. ¡Lee los documentos!
dt = DecisionTreeClassifier(max_depth=2).fit(X_train, y_train)
dt.score(X_test, y_test)
```

```python
# cantidad normalizada que cada característica contribuyó a la puntuación
dt.feature_importances_
```

```python
two_most_important_features = dt.feature_importances_.argsort()[-2:]; two_most_important_features
```
